## Pumpkin Pricing

Load up required libraries and dataset. Convert the data to a dataframe containing a subset of the data: 

- Only get pumpkins priced by the bushel
- Convert the date to a month
- Calculate the price to be an average of high and low prices
- Convert the price to reflect the pricing by bushel quantity

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime

pumpkins = pd.read_csv('../data/US-pumpkins.csv')

pumpkins.head()
```

```python
pumpkins = pumpkins[pumpkins['Package'].str.contains('bushel', case=True, regex=True)]

columns_to_select = ['Package', 'Variety', 'City Name', 'Low Price', 'High Price', 'Date']
pumpkins = pumpkins.loc[:, columns_to_select]

price = (pumpkins['Low Price'] + pumpkins['High Price']) / 2

month = pd.DatetimeIndex(pumpkins['Date']).month
day_of_year = pd.to_datetime(pumpkins['Date']).apply(lambda dt: (dt-datetime(dt.year,1,1)).days)

new_pumpkins = pd.DataFrame(
    {'Month': month, 
     'DayOfYear' : day_of_year, 
     'Variety': pumpkins['Variety'], 
     'City': pumpkins['City Name'], 
     'Package': pumpkins['Package'], 
     'Low Price': pumpkins['Low Price'],
     'High Price': pumpkins['High Price'], 
     'Price': price})

new_pumpkins.loc[new_pumpkins['Package'].str.contains('1 1/9'), 'Price'] = price/1.1
new_pumpkins.loc[new_pumpkins['Package'].str.contains('1/2'), 'Price'] = price*2

new_pumpkins.head()
```

# Regression models at work
## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
## Linear Regression Model
X = pie_pumpkins['DayOfYear'].to_numpy().reshape(-1,1)
y = pie_pumpkins['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
pred = lin_reg.predict(X_test)
```

```python
mse = np.sqrt(mean_squared_error(y_test, pred))
print(mse)
print(f'Mean error: {mse:3.3} ({mse/np.mean(pred)*100:3.3}%)')
```

```python
score = lin_reg.score(X_train, y_train)
print('Model determination: ', score)
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, pred)
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

pipeline = make_pipeline(PolynomialFeatures(2), LinearRegression())
pipeline.fit(X_train, y_train)

pred = pipeline.predict(X_test)

plt.scatter(X_test, y_test)
plt.plot(sorted(X_test), pipeline.predict(sorted(X_test)))
```

```python
mse = np.sqrt(mean_squared_error(y_test, pred))
print(mse/np.mean(pred))
```

```python
pipeline3 = make_pipeline(PolynomialFeatures(3), LinearRegression())
pipeline3.fit(X_train, y_train)

pred3 = pipeline3.predict(X_test)

plt.scatter(X_test, y_test)
plt.plot(sorted(X_test), pipeline3.predict(sorted(X_test)))
```

```python
mse = np.sqrt(mean_squared_error(y_test, pred3))
print(mse/np.mean(pred3))
```

```python
pipeline4 = make_pipeline(PolynomialFeatures(4), LinearRegression())
pipeline4.fit(X_train, y_train)

pred4 = pipeline4.predict(X_test)

plt.scatter(X_test, y_test)
plt.plot(sorted(X_test), pipeline4.predict(sorted(X_test)))
```

```python
mse = np.sqrt(mean_squared_error(y_test, pred4))
mse/np.mean(pred4)
```

```python
## Create an X to include all the data

X = pd.get_dummies(new_pumpkins['Variety']) \
 .join(new_pumpkins['Month']) \
 .join(pd.get_dummies(new_pumpkins['City'])) \
 .join(pd.get_dummies(new_pumpkins['Package']))
y = new_pumpkins[('Price')]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### With a simple linear regression ...

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
mse = np.sqrt(mean_squared_error(y_test, pred))
score = model.score(X_train, y_train)
print('Model determination: ', score)
print('Mean Squared Error: ', mse/np.mean(pred))
```
